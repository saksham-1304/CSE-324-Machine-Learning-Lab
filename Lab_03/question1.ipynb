```python
# ============================================
# Mall Customer Segmentation using K-Means
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# -------------------------------
# 1. Load Dataset
# -------------------------------
df = pd.read_csv("Mall_Customers.csv")

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
# -------------------------------
# 2. Select Features
# -------------------------------
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# -------------------------------
# 3. Scaling
# -------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# -------------------------------
# 4. Elbow Method
# -------------------------------
wcss = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    print(kmeans)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
plt.plot(range(2,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# -------------------------------
# 5. Apply Optimal K-Means
# (Usually k=5 for this dataset)
# -------------------------------
k_optimal = 5
kmeans = KMeans(n_clusters=k_optimal, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

df['Cluster'] = labels
```

```python
df.head()
```

```python
# -------------------------------
# 6. Silhouette Score
# -------------------------------
score = silhouette_score(X_scaled, labels)
print("Silhouette Score:", score)
```

```python

# -------------------------------
# 7. Plot Clusters
# -------------------------------
plt.figure(figsize=(8,6))
sns.scatterplot(x=X.iloc[:,0], y=X.iloc[:,1], hue=labels, palette='Set2')

# Plot centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:,0], centroids[:,1], c='black', s=200, marker='X', label='Centroids')

plt.title("Customer Segments")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python
# -------------------------------
# 8. Business Interpretation
# -------------------------------
for i in range(k_optimal):
    print(f"\nCluster {i}")
    print(df[df['Cluster']==i][['Annual Income (k$)','Spending Score (1-100)']].mean())
```
